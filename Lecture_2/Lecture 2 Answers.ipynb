Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

Step 1: To determine the expected value $E[y]$, we employ the formula:

$$E[y] = \Sigma y \cdotp P(y)$$

Step 2: With the joint probability distribution given as $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$, we substitute it into the formula:

$$E[y] = \Sigma y \cdotp \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$$

Step 3: Given $Y_1 +Y_2 +Y_3 +···+Y_k =n$, and since the trials are independent, we can rewrite the joint probability as:

$$E[y] = \sum (y_1 \cdot P_1 + y_2 \cdot P_2 + y_3 \cdot P_3 + \ldots + y_n \cdot P_n)$$
$$E[y] = \sum y_i \cdot P_i  $$

Step 4: Since $P_i = p_i$ (the probability of outcome falling into cell $i$), we obtain: 

$$E[y] = \sum y_i \cdot p_i  $$
$$E[y] = n \cdot p_i  $$

$\textbf{Conclusion:}$ This demonstrates that the expected value of the random variable $y$ is $n \cdot p_i $

If 500 adults are sampled  randomly, find the probability rhat the sample contains 100 person between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?

To find the probability that the sample contains 100 persons between the ages of 18 and 24, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64, we'll make use of the multinomial probability formula:

$P(y) = \frac{n!}{y_1!y_2!y_3!} P_1^{y_1} P_2^{y_2} P_3^{y_3}$

Given:
- $n = 500$ (total number of adults sampled)
- $k = 5$ (number of age groups)
- $p_1 = 0.18$, $p_2 = 0.23$, $p_3 = 0.16$, $p_4 = 0.27$, $p_5 = 0.16$ (proportions for each age group)

The probability mass function (PMF) for a multinomial distribution is:

$P(y) = \frac{n!}{y_1!y_2!y_3!y_4!y_5!} p_1^{y_1} p_2^{y_2} p_3^{y_3} p_4^{y_4} p_5^{y_5}$

Where $y_1$, $y_2$, $y_3$, $y_4$, $y_5$ are the counts for each age group.

- $y_1 = 100$ (for ages 18-24)
- $y_2 = 200$ (for ages 25-34)
- $y_3 = 0$ (no persons between ages 35-44)
- $y_4 = 200$ (for ages 45-64)
- $y_5 = 500 - (y_1 + y_2 + y_3 + y_4) = 500 - (100 + 200 + 0 + 200) = 0$ (for ages 65-100)

Calculate by substituting these values into the PMF:

$P(y) = \frac{500!}{100!200!0!200!0!} (0.18)^{100} (0.23)^{200} (0.16)^{0} (0.27)^{200} (0.16)^{0}$

$P(y) = \frac{500!}{100!200!200!} (0.18)^{100} (0.23)^{200} (0.27)^{200}$

The expected value to obtain a person in the 65 and above age group is given by the sum of the products of each outcome and its probability:

$E[Y_5] = n \times p_5 = 500 \times 0.16 = 80$

Therefore, we conclude that the expected value to obtain a person aged 65 and 100 given a sample size of 500 adults is 80.

Task 3: Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Step 1: Begin with the equation for the expected value $E(y)$:
$$[E(y) = e^{-\lambda} \left( \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!} + 1 \right)$$

Step 2: Recognize that the summation $\sum_{k=0}^{\infty} \frac{\lambda^k}{k!}$ represents the Taylor series expansion of $(e^\lambda)$, and substitute it into the equation:
$$E(y) = e^{-\lambda} (\lambda e^\lambda + 1)$$

Step 3: Apply the distribution of $(e^{-\lambda})$ into the parentheses:
$$E(y) = \lambda e^{-\lambda} e^\lambda + e^{-\lambda} $$
$$E(y) = \lambda \cdot 1 + e^{-\lambda} $$
$$E(y) = \lambda + e^{-\lambda} $$

In the final expression, $e^{-\lambda}$ serves as a constant factor independent of $y$, thereby not influencing the expected value. Thus, the expected value of the distribution simplifies to $\lambda$.

$$
P(Y = y) = \frac{e^{-\lambda} \lambda^y}{y!}

$$
P(Y = 0) = \frac{e^{-5} 5^0}{0!} = e^{-5}
$$

$$
