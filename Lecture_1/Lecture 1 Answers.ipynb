1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$

Step 1: Let's start by supposing that $e$ can be expressed as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers with no common factors except 1, and $q \neq 0$.

Step 2: Remember that $e$ is defined as the sum of an infinite series:
 
$ e = \Sigma_{k=0}^\infty \frac{1}{k!}$

Step 3: Let's examine the partial sums of this series. We'll denote the $n$th partial sum as $S_n$:

$S_n = \Sigma_{k=0}^\infty \frac{1}{k!}$

Step 4: We define $e$ as the limit of these partial sums as $n$ tends to infinity:

$e = \lim_{n \to \infty} S_n$

Step 5: If $e$ is rational, then the partial sums $S_n$ must also tend towards a rational number as $n$ increases infinitely.

Step 6: Note that each term in the series $\frac{1}{k!}$ is positive, so each partial sum $S_n$ is increasing.

Step 7: Let's consider the factorial terms $k!$. For any $k \geq 1$, $k!$ is always greater than $k$, which means $\frac{1}{k!}$ is less than $\frac{1}{k}$.

Step 8: Recognize that the series $\sum_{k=0}^\infty \frac{1}{k}$ diverges (known as the harmonic series).

Step 9: Since $\frac{1}{k!}$ is smaller than $\frac{1}{k}$, we conclude that the series $\sum_{k=0}^\infty \frac{1}{k!}$ converges.

Step 10: If the partial sums $S_n$ converge to a rational number, they must be bounded.

Step 11: However, the convergence of $\sum_{k=0}^\infty \frac{1}{k!}$ contradicts the unboundedness of the harmonic series.

$\textbf{Conclusion}$: Hence, assuming $e$ is rational leads to a contradiction. Thus, we conclude that $e$ is not rational, i.e., $e \notin \mathbb{Q}$.

2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:

\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}

\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}

Let's demonstrate Minkowski's Inequality for sums directly using the Triangle Inequality theorem:

$$
|a + b| \leq |a| + |b|
$$

Step 1: We begin by exponentiating both sides of the inequality to the power of 
p:

$$
|a + b|^{p} \leq |a|^{p} + |b|^{p}  
$$

Step 2: Introduce summation notation for all k from 1 to n:

$$
\sum_{k = 1}^{n} |a_k + b_k|^{p} \leq \sum_{k=1}^{n} |a_k|^{p} + \sum_{k =1 }^{n} |b_k|^{p} 
$$

Step 3: Raise to 1 over p by property of absolute value

$$
[\sum_{k = 1}^{n}|a_k + b_k|^{p}]^{\frac{1}{p}} \leq [\sum_{k = 1}^{n} |a_k|^{p}]^{\frac{1}{p}} + [\sum_{k = 1}^{n}|b_k|^{p}]^{\frac{1}{p}}
$$

Hence, we have proved Minkowski's Inequality for sums by utilizing the Triangle Inequality theorem.
