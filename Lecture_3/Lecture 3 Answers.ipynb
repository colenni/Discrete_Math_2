$\textbf{Graphs}$
-

- A graph $G$ consists of two finite sets: a nonempty set $V(G)$ of vertices and a set $E(G)$ of edges, where each edge is associated with a set consisting of either one or two vertices called its endpoints. The correspondence from edges to endpoints is called the edge-endpoint function.

- An edge with just one endpoint is called a loop, and two or more distinct edges with the same set of endpoints are said to be parallel. An edge is said to connect its endpoints; two vertices that are connected by an edge are called adjacent; and a vertex that is an endpoint of a loop is said to be adjacent to itself.

- An edge is said to be incident on each of its endpoints, and two edges incident on the same endpoint are called adjacent. A vertex on which no edges are incident is called isolated.

```python
import matplotlib.pyplot as plt
import networkx as nx

n = 8  # Number of nodes
m = 5  # Number of edges
seed = 55  # Seed for random number generator

# Use seed for reproducibility
G = nx.gnm_random_graph(n, m, seed=seed)

# Add self-loops to the remaining nodes
edgelist = [(0, 0), (7, 7)]
G.add_edges_from(edgelist)

# Circular layout
pos = nx.circular_layout(G)

# Draw the graph with different orientations
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```

$\textbf{Directed Graph (Digraphs)}$
-

A directed graph, or digraph, consists of two finite sets: a nonempty set V (G) of vertices and a set D(G) of directed edges, where each is associated with an ordered pair of vertices called its endpoints. If edge e is associated with the pair (v, w) of vertices, then e is said to be the (directed) edge from v to w.

The figure depicted above illustrates a directed graph, also known as a digraph. $\textbf{Bipartite Graphs}$
-

- Let $m$ and $n$ be positive integers. A complete bipartite graph on $(m, n)$ vertices, denoted $K_{m,n}$, is a simple graph with distinct vertices $v_1, v_2, . . . , v_m$ and $w_1, w_2, . . . , w_n$ that satisfies the following properties: $\forall i,k = 1,2,...,m$ and $\forall j,l = 1,2,...,n$

1. There is an edge from each vertex $v_i$ to each vertex $w_j$.
2. There is no edge from any vertex $v_i$ to any other vertex $v_k$.
3. There is no edge from any vertex $w_j$ to any other vertex $w_l$.



```python
import matplotlib.pyplot as plt
import networkx as nx

kabit_network = nx.Graph()
kabit_network.add_nodes_from([1,3,5], bipartite=0) # Add the node attribute "bipartite"
kabit_network.add_nodes_from([2,4], bipartite=1)
kabit_network.add_edges_from([(1,2), (1,4)
                              , (3,2), (3,4)
                              , (5,2), (5,4)])

top = nx.bipartite.sets(kabit_network)[0]
pos = nx.bipartite_layout(kabit_network, top)
nx.draw(kabit_network, pos=pos, with_labels=True, node_color=['yellow','yellow','yellow','yellow','yellow'])
plt.show()
```

$\textbf{Subgraphs}$
-

A graph $H$ is said to be a subgraph of a graph $G$ if, and only if, every vertex in $H$ is also a vertex in $G$,every edge in $H$ is also an edge in $G$,and every edge in $H$ has the same endpoints as it has in $G$.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the main graph
G_main = nx.Graph()
G_main.add_nodes_from([1, 2, 3, 4])
G_main.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Create the subgraph (a subset of nodes and edges from the main graph)
subgraph_nodes = [1, 2, 3]
subgraph_edges = [(1, 2), (2, 3)]
G_sub = nx.Graph()
G_sub.add_nodes_from(subgraph_nodes)
G_sub.add_edges_from(subgraph_edges)

# Draw the main graph
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G_main, with_labels=True, node_color='red', node_size=500, font_size=12, font_color='black', width=1)
plt.title("Main Graph")

# Draw the subgraph
plt.subplot(122)
nx.draw(G_sub, with_labels=True, node_color='violet', node_size=500, font_size=12, font_color='black', width=1)
plt.title("Subgraph")

plt.tight_layout()
plt.show()
```

$\textbf{Degree of a Vertex}$
-

Let $G$ be a graph and $v$ a vertex of $G$. The degree of $v$, denoted $deg(v)$, equals the number of edges that are incident on $v$, with an edge that is a loop counted twice. The total degree of $G$ is the sum of the degrees of all the vertices of $G$.

$\textbf{The Handshake Theorem}$

If $G$ is any graph, then the sum of the degrees of all the vertices of $G$ equals twice the number of edges of $G$. Specifically, if the vertices of $G$ are $v_1,v_2,...,v_n$, where $n$ is a nonnegative integer, then
the total degree of $G = deg(v_1) + deg(v_2 ) +    + deg(v_n )$ = 2$\cdotp$(the number of edges of $G$).

The Handshake theorem can be used to determine if a Graph exists because the total degree should always be even.

$\textbf{Travelling in a Graph}$
-

- Travelling in a graph is accomplished by moving from one vertex to another along a sequence of adjaceny edges.

- There are different types of travelling in a graph, consider a graph $G$, and let $w$ and $w$ be vertices in $G$, then:
    - A $\textbf{walk from v to w}$ is a finite alternative sequence of adjacent vertices and edges of $G$. The walk has the form: $v_0 e_1 v_1 e_2 v_2 .... e_{n-1} v_{n-1} e_n v_n$.
    - A $\textbf{trail from v to w}$ is a walk  from $v$ to $w$ that does not contain a repeated edge.
    - A $\textbf{path from v to w}$ is a walk  from $v$ to $w$ that does not contain a repeated vertex.
    - A $\textbf{closed walk}$ is a walk  from $v$ to $w$ that starts and ends in the same vertex.
    - A $\textbf{circuit}$ is a closed walk that contains at least one edge and does not contain a repeated edge.
    - A $\textbf{simple circuit}$ is a circuit that does not have any other repeated except the first and the last.

$\textbf{Trail}$

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add edges to the graph representing streets between intersections
streets = [(1, 2), (1, 3), (2, 4), (3, 4), (3, 5), (4, 5), (4, 6), (5, 7), (6, 7)]

# Add edges to the graph
G.add_edges_from(streets)

# Function to find a trail using Depth-First Search
def dfs_trail(graph, start, visited=None, trail=None):
    if visited is None:
        visited = set()
    if trail is None:
        trail = []

    visited.add(start)
    trail.append(start)

    for neighbor in graph.neighbors(start):
        if neighbor not in visited:
            dfs_trail(graph, neighbor, visited, trail)

    return trail

# Find a trail starting from intersection 1
trail = dfs_trail(G, 1)

# Draw the graph
pos = nx.spring_layout(G)  # Layout for visualization
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold', node_size=800)
plt.show()

print("Trail:", trail)
```

$\textbf{Connectedness a Graphs}$
-

- Let $G$ be a graph, and $v$ and $w$ be vertices in $G$. The vertices $v$ and $w$ are connected $\iff$ there is a walk from $v$ to $w$.
- G is a $\textbf{connected graph} \iff$ given any two vertices $(v,w)\in G$, there is a walk from $v$ to $w$.
- G is connected $\iff \forall v,w \in V(G), \exists$ a walk from $v \to w$.
- If $G$ is connected, then any two distinct vertices of $G$ can be connected by a path.
- If vertices $v$ and $w$ are part of a circuit in $G$ and one edge is removed from the circuit, then there still exists a trail from $v$ to $w$ in $G$.
- If $G$ is connected and $G$ contains a circuit, then an edge of the circuit can be removed without disconnecting $G$.

```python
import matplotlib.pyplot as plt
import networkx as nx

mySeed = 50

node_names = ['A', 'B', 'C', 'D', 'E']
name_map = {i: name for i, name in enumerate(node_names)}
G = nx.gnm_random_graph(len(node_names), 5, seed=mySeed)
G = nx.relabel_nodes(G, name_map)

pos = nx.spring_layout(G, seed=mySeed)
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=500)
```

$\textbf{Euler's Circuit}$
-

- Let $G$ be a graph, a $\textbf{Euler's circuit}$ that contains every vertex and every edge of G.
- The $\textbf{Euler's circuit}$ for $G$ is a sequence of adjacent vertices, and edges in $G$ that has at least one edge; starts and ends with the same vertex. Uses every vertex of G at least once, and uses every edge of $G$ $\textbf{"EXACTLY ONCE"}$.
- If $G$ is connected, and the degree of every vertex of $G$ is a positive even integer, then G has a $\textbf{Euler's circuit}$.
- $G$ has a $\textbf{Euler's circuit} \iff$ $G$ is connected and every vertex of $G$ has positive even degree.

$\textbf{Euler's Trail}$
-

- Let $G$ be a graph, and $v$ and $w$ be distinct vertices in $G$, then a $\textbf{Euler's Trail}$ from $v$ to $w$ is a sequence of adjacent edges and vertices that starts at $v$ and ends at $w$, passes through every vertex of $G$ at least once, and traverses every edge of $G$ $\textbf{"EXACTLY ONCE"}$.
- If $v$ and $w$ are two distinct vertices of $G$, then there is a $\textbf{Euler's path}$ from $v$ to $w \iff G$ is connected; $v$ and $w$ have odd degree, and all other vertices of $G$ has a positive even degree. 

```python
import matplotlib.pyplot as plt
import networkx as nx
plt.figure(figsize=(20, 5))

G = nx.Graph()
G.add_nodes_from([1,2,3,4,5])
Gnew = nx.Graph()
Gnew.add_nodes_from([1,2,3,4])
edgelist = [(1,2),(2,3),(3,1),(4,5),(1,4),(2,4),(2,5),(3,4),(1,5),(3,5)]
edgelist2 = [(1,2),(2,4),(4,3),(3,2)]

plt.subplot(121)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=edgelist, edge_color='green')
plt.title('Eulers Circuit')

plt.subplot(122)
nx.draw(Gnew, pos=nx.circular_layout(Gnew), with_labels=True)
nx.draw_networkx_edges(Gnew, pos=nx.circular_layout(Gnew), edgelist=edgelist2, edge_color='yellow')
plt.title('Eulers Trail')

plt.show()
```

$\textbf{Hamiltonian Circuits}$
-

- A Hamiltonian circuit for G is a sequence of adjacent vertices and distinct edges in which every vertex of G appears exactly once, except for the first and the last, which are the same.

If a graph $G$ has a Hamiltonian circuit, then $G$ has a subgraph $H$ with the following properties:
1. $H$ contains every vertex of $G$.
2. $H$ is connected.
3. $H$ has the same number of edges as vertices.
4. Every vertex of $H$ has degree 2.

```python
import matplotlib.pyplot as plt
import networkx as nx
plt.figure(figsize=(20, 5))

G = nx.Graph()
G.add_nodes_from([1,2,3,4,5])
Gnew = nx.Graph()
Gnew.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(4,5),(1,4),(2,4),(2,5),(3,4),(1,5),(3,5)]
edgelist2 = [(1,2),(2,3),(3,4),(4,5),(5,1)]

plt.subplot(121)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=edgelist)
plt.title('Original Graph')

plt.subplot(122)
nx.draw(Gnew, pos=nx.circular_layout(Gnew), with_labels=True)
nx.draw_networkx_edges(Gnew, pos=nx.circular_layout(Gnew), edgelist=edgelist2, edge_color='orange')
plt.title('Hamiltonian Circuit')

plt.show()
```

$\textbf{The Travelling Salesman Problem}$
---

The general traveling salesman problem involves finding a Hamiltonian circuit to minimize the total distance traveled for an arbitrary graph with n vertices in which each edge is marked with a distance.

- At present, there is no known algorithm for solving the general traveling salesman problem that is more efficient. However, there are efficient algorithms that find pretty good solutionsthat is, circuits that, while not necessarily having the least possible total distances, have smaller total distances than most other Hamiltonian circuits.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Use seed for reproducibility
seed = 55

# Create a graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D"])

# Add edges with weights
edgelist = [("A", "B", 30), ("A", "C", 50), ("A", "D", 40), ("B", "C", 30), ("B", "D", 35), ("C", "D", 25)]
G.add_weighted_edges_from(edgelist)

# Define layout
pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout

# Draw edges with labels
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): '10',
                 ('A', 'C'): '30',
                 ('A', 'D'): '40',
                 ('B', 'C'): '10',
                 ('B', 'D'): '35',
                 ('C', 'D'): '15'},
    font_color='red'
)

# Draw the graph
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='violet',
    with_labels=True
)

plt.show()
```
